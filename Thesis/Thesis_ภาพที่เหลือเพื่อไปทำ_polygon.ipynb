```python
original_U2M = original[original['Filename'].isin(pos_U2M['filename'])]
print(original_U2M.shape)
original_U2M.head(5)
```

```python
done = original_U2M[(original_U2M['Age(year)']>=10)&(original_U2M['Age(year)']<=18)]
```

```python
other = original_U2M[(original_U2M['Age_predict_int']>=9.0)&(original_U2M['Age_predict_int']<=25.0)]
```

```python
# prompt: select only row  filemane in other not repreat with done

# Assuming 'original_U2M' and 'done' DataFrames are already defined as in the provided code.

# Find filenames in 'other' that are NOT in 'done'
other_unique = other[~other['Filename'].isin(done['Filename'])]

# Now 'other_unique' contains only the rows from 'other' where the 'Filename' is not present in 'done'
other_unique
```

```python
from PIL import Image   # PIL = module ที่ใช้ในการจัดการภาพ
import matplotlib.pyplot as plt # แสดงภาพ

path = other_unique['Path_Name'].tolist()
for i in path:
    filename = i.split("/")[-1]
    img = Image.open(i) #Image.open(i) คือ คำสั่งที่ใช้โหลดรูปภาพจาก path หรือที่อยู่ของไฟล์รูปภาพที่ระบุไว้ในตัวแปร i โดย Image เป็น module ที่ใช้จัดการรูปภาพและสร้างเครื่องมือในการประมวลผลภาพต่างๆ ซึ่งต้องติดตั้งก่อนใช้งาน และ open เป็น method ที่ใช้เปิดไฟล์ภาพ ซึ่งจะอ่านข้อมูลภาพและโหลดภาพให้เรียบร้อยก่อนที่จะนำไปใช้งานต่อไป
    # print(i)
    img.save('/content/gdrive/MyDrive/DentDiscovering/UpperSecondMolar/Images/other/'+filename)
```
