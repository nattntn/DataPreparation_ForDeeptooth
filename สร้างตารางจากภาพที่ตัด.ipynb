# สร้างตารางภาพที่ตัดแล้ว เฉพาะ Normal

```python
paro = r'/content/gdrive/MyDrive/Tooth_Shap_GPT/Deep_tooth/Data_Cut_images/Normal' # ให้อ่านที่อยู่ของโฟรเดอร์ในไดร์ฟ และสร้างตัวแปร ชื่อว่า paro7 มารับไว้
```

```python
import os

files = []  # สร้าง list เพื่อเก็บตำแหน่งของไฟล์
for path, dirnames, filenames in os.walk(paro): # os.walk คือ ทำงานสร้างชื่อไฟล์แต่ละไฟล์ ตามที่เก็บข้อมูล
    # ตรวจสอบว่าโฟลเดอร์ที่ชื่อ "Unflip" อยู่ใน dirnames หรือไม่
    if 'Unflip' in dirnames:
        dirnames.remove('Unflip')  # ข้ามการค้นหาในโฟลเดอร์ "Unflip"

    # ตรวจสอบว่าโฟลเดอร์ว่างหรือไม่ (ไม่มีไฟล์ในโฟลเดอร์)
    if not filenames:
        continue  # ข้ามการค้นหาในโฟลเดอร์ที่ไม่มีไฟล์

    files.extend(os.path.join(path, name) for name in filenames) # เพิ่มตำแหน่งของไฟล์ลงใน list ถ้าเป็นไฟล์ในโฟลเดอร์ที่ต้องการ # extend  ใช้เพื่อรวม list # คำสั่งเชื่อมส่วนประกอบของ path ที่ประกอบด้วย directory หรือ file เพื่อสร้าง path หรือ URL ที่รองรับการทำงานบนระบบปฏิบัติการต่างๆ

print(files)
```
