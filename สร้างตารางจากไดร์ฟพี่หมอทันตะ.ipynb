```python
paro = r'/content/gdrive/MyDrive/Film pano' # ให้อ่านที่อยู่ของโฟรเดอร์ในไดร์ฟ และสร้างตัวแปร ชื่อว่า paro7 มารับไว้
```

```python
list_path = list(pathname['Path_Name']) # เก็บที่อยู่ของภาพในรูปแบบตาราง โดยชื่อว่า list_path ซึ่ง list(ชื่อตาราง['ชื่อคอลัมน์'])
for i in list_path: # วนลูปใน list_path
    res = i.split('/') # split ใช้สำหรับแยก โดยในที่นี้แยกด้วย / แล้วตั้งชื่อว่า res
    print(res) # แสดงผลลัพธ์ของ res ตัวที่ 7 คือตำเเหน่งที่เราต้องการ
```

```python
list_path = list(pathname['Path_Name']) # เก็บที่อยู่ของภาพในรูปแบบตาราง โดยชื่อว่า list_path ซึ่ง list(ชื่อตาราง['ชื่อคอลัมน์'])
class_year = list() # สร้างตัวแปร class_list มาเก็บ list ว่าง
class_gender = list() # สร้างตัวแปร class_list มาเก็บ list ว่าง
class_pic = list() # สร้างตัวแปร class_list มาเก็บ list ว่าง
class_folder_nor = list() # สร้างตัวแปร class_list มาเก็บ list ว่าง
class_folder_abnor = list()
for i in list_path: # วนลูปใน list_path
    res = i.split('/') # split ใช้สำหรับแยก โดยในที่นี้แยกด้วย / แล้วตั้งชื่อว่า res
    print(res) # แสดงผลลัพธ์ของ res ตัวที่ 7 คือตำเเหน่งที่เราต้องการ
    class_year.append(res[5]) # append  คำสั่งเพิ่มข้อมูลลงใน List
    class_gender.append(res[6]) # append  คำสั่งเพิ่มข้อมูลลงใน List
    class_pic.append(res[-1]) # append  คำสั่งเพิ่มข้อมูลลงใน List
    class_folder_nor.append(res[-2]) # append  คำสั่งเพิ่มข้อมูลลงใน List
    class_folder_abnor.append(res[-3])

pathname['Age(year)'] = class_year # ตั้งชื่อคอลัมน์ว่า Part ในตาราง pathname โดยนำค่าใน class_list มาใส่
pathname['Gender'] = class_gender
pathname['Filename'] = class_pic
pathname['Folder_normal'] = class_folder_nor
pathname['Folder_abnormal'] = class_folder_abnor
```
